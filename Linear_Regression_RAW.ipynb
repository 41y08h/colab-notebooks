```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("moore.csv", index_col=False)
data
```

```python
plt.scatter(data["1971"], data["2300"]);
```

```python
X = data["1971"] - np.mean(data["1971"])
y = np.log(data["2300"])

plt.scatter(X, y);
```

```python
def gradient_descent(X, y, m=0, b=0):
  m_gradient = 0;
  b_gradient = 0;

  n = len(X)

  total_loss = 0
  for i in range(n):
    loss = y[i] - m*X[i] - b

    m_gradient += (-2/n) * (loss) * X[i]
    b_gradient += (-2/n) * (loss)
    total_loss += loss
 
  return (m_gradient, b_gradient, total_loss)
```

```python
m = 0
b = 0
epochs = 10000
L = 0.001

for i in range(epochs):
  m_gradient, b_gradient, loss = gradient_descent(X, y, m, b)

  m = m - m_gradient*L
  b = b - b_gradient*L

  print("Epoch :{} / Loss: {}".format(i+1, loss))

print(m, b)
```

```python
plt.scatter(X, y)

X_sample = np.linspace(-30, 15)
plt.plot(X_sample, (m*X_sample + b), color="black");
```
